```python
# Lm sch tn ct  trnh li vi Snowflake
def clean_column_names(df):
    """Lm sch tn ct  tng thch vi Snowflake"""
    # To mt bn sao  khng thay i DataFrame gc
    cleaned_df = df.copy()

    # Lm sch tn ct
    new_columns = []
    for col in cleaned_df.columns:
        # Loi b k t c bit, chuyn thnh ch hoa
        clean_col = str(col).strip().upper()
        # Thay th khong trng v k t c bit bng underscore
        import re
        clean_col = re.sub(r'[^A-Z0-9_]', '_', clean_col)
        # Loi b underscore lin tip
        clean_col = re.sub(r'_+', '_', clean_col)
        # Loi b underscore  u v cui
        clean_col = clean_col.strip('_')
        new_columns.append(clean_col)

    cleaned_df.columns = new_columns
    return cleaned_df

# p dng lm sch tn ct
df_clean = clean_column_names(df)
print("Tn ct sau khi lm sch:")
print(df_clean.columns.tolist())
print("\nKim tra vi dng u ca d liu  lm sch:")
df_clean.head()
```
